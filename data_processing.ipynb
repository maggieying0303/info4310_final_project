```python
print("average ratings per platform")
print("disney movie average ratings", disney_movies.mean(axis=0, skipna=True)['IMDb'])
print("hulu movie average ratings", hulu_movies.mean(axis=0, skipna=True)['IMDb'])
print("netflix movie average ratings", netflix_movies.mean(axis=0, skipna=True)['IMDb'])
print("prime movie average ratings", prime_movies.mean(axis=0, skipna=True)['IMDb'])
```

```python
g_list = ['Drama', 'Comedy', 'Thriller', 'Action', 'Documentary', 'Family', 'Adventure', 'Fantasy']
dict = {'Drama':[], 'Comedy':[], 'Thriller':[], 'Action':[], 'Documentary':[], 'Family':[], 'Adventure':[], 'Fantasy':[]}
for g in g_list:
    print("average ratings for "+ g + " movies per platform")

    print("")

    disney_copy = disney_movies.copy().reset_index()

    for i in np.arange(len(disney_copy.index)):
        if type(disney_copy['Genres'][i]) != str or g not in disney_copy['Genres'][i]:
            disney_copy = disney_copy.drop(i)

    mean = disney_copy.mean(axis=0, skipna=True)['IMDb']
    print("Disney "+ g + " movie average ratings", mean)
    dict[g].append(mean)

    hulu_copy = hulu_movies.copy().reset_index()

    for i in np.arange(len(hulu_copy.index)):
        if type(hulu_copy['Genres'][i]) != str or g not in hulu_copy['Genres'][i]:
                hulu_copy = hulu_copy.drop(i)
 
    mean = hulu_copy.mean(axis=0, skipna=True)['IMDb']
    print("Hulu "+ g + " movie average ratings", mean)
    dict[g].append(mean)
    
    netflix_copy = netflix_movies.copy().reset_index()

    for i in np.arange(len(netflix_copy.index)):
        if type(netflix_copy['Genres'][i]) != str or g not in netflix_copy['Genres'][i]:
                netflix_copy = netflix_copy.drop(i)

    mean = netflix_copy.mean(axis=0, skipna=True)['IMDb']
    print("Netflix "+ g + " movie average ratings", mean)
    dict[g].append(mean)
                
    prime_copy = prime_movies.copy().reset_index()

    for i in np.arange(len(prime_copy.index)):
        if type(prime_copy['Genres'][i]) != str or g not in prime_copy['Genres'][i]:
                prime_copy = prime_copy.drop(i)

    mean = prime_copy.mean(axis=0, skipna=True)['IMDb']
    print("Prime "+ g + " movie average ratings", mean)
    dict[g].append(mean)
    
    print("")
```

```python
df = pd.DataFrame.from_dict(dict)
df.index = ["Disney", "Hulu", "Netflix", "Prime"]
df.to_csv("genre_ratings.csv")
```
